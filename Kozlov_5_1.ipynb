Задание размеров матриц и коэффициентов  

```python
n = 100
a = b = 10

A = np.zeros((n, n))
C = np.zeros((n, 1))

```

Заполнение матриц

```python
for i in range(n):
    C[i][0] = i + 1

for i in range(n):
    for j in range(n):
        if i == j:
            A[i][j] = a
        if j == i + 2:
            A[i][j] = 1/b
        if j == i - 1 or j == i + 1:
            A[i][j] = 1
```

Вывод матрицы коэффициентов

```python
print("Матрица коэффициентов уравнений: \n ", A)
```

Метод Гаусса

```python
def Gauss(A, C):
    
    n = A.shape[0]
    
    # создаем расширенную матрицу системы
    AC = np.hstack((A, C.reshape(n, 1)))
    print("Расширенная матрица имеет вид: \n", AC)
    
    # прямой ход метода Гаусса
    for i in range(n):
        # выбираем максимальный по модулю элемент в столбце i
        max_el_row = np.argmax(np.abs(AC[i:, i])) + i

        # меняем местами строки i и max_el_row
        AC[[i, max_el_row]] = AC[[max_el_row, i]]

        # приводим матрицу к треугольному виду
        for j in range(i + 1, n):
            coeff = AC[j, i] / AC[i, i]
            AC[j] -= coeff * AC[i]

    # обратный ход метода Гаусса
    X = np.zeros(n)
    for i in range(n - 1, -1, -1):
        X[i] = (AC[i, n] - AC[i, i + 1:n] @ X[i + 1:n]) / AC[i, i]

    print("Решение системы уравнений:\n", X)
```

Вызов функции, реализующей метод Гаусса

```python
print(Gauss(A, C))
```

Метод Якоби

```python
# метод Якоби решения СЛАУ
def Yakobi(A, C):
    # Начальное приближение решения
    X = np.zeros((n, 1))

    # Максимальное число итераций
    max_iterations = 1000

    # Критерий сходимости
    epsilon = 1e-6

    # Матрицы D, L и U
    D = np.diag(np.diag(A)) # Диагональные элнмннты
    L = -np.tril(A, k=-1) # Нижний треугольник
    U = -np.triu(A, k=1) # Верхний треугольник

    # Итерационный процесс метода Якоби
    for iteration in range(max_iterations):
        X_new = np.linalg.inv(D).dot(C + (L + U).dot(X))

        # Проверка критерия сходимости
        if np.linalg.norm(X_new - X) < epsilon:
            break

        X = X_new
        X_ = X.flatten()
    return X_
```

Вызов функции, реализующей метод Якоби

```python
# Результат метода Якоби
print("Решение системы уравнений методом Якоби \n", Yakobi(A, C))
```

Проверка правильности решения через функцию solve 

```python
# Проверка через solve
Y = np.linalg.solve(A, C)
# преобразование двумерного массива решений в одномерный массив
Y_ = Y.flatten()
print("Решение системы уравнений через solve \n", Y_)
```

Вычисление максимального и минимального собственного значений матрицы и их вывод

```python
eig_vals, _ = np.linalg.eig(A)
l_max = max(eig_vals)
l_min = min(eig_vals)

print("\nМаксимальное собственное значение: ", l_max)
print("Минимальное собственное значение: ", l_min)
```

Вычисление норм матрицы A и обратной ей 

```python
# Вычисление норм матрицы A и обратной ей
norm_A = np.linalg.norm(A)
print(f"\nНорма матрицы А: {norm_A}")


A_inv = np.linalg.inv(A)
norm_A_inv = np.linalg.norm(A_inv)
print(f"Норма обратной матрицы матрицы А: {norm_A_inv}")
```

Вычисление числа обусловленности матрицы

```python
# Вычисление числа обусловленности матрицы A
mu = norm_A * norm_A_inv
print(f"Число обусловленности матрицы A = {mu}")
```
