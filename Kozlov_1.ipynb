```python
import numpy as np
import matplotlib.pyplot as plt
```

Задаём шаг и диапазон

```python
h=0.1
x = np.arange(-5, 5, h)
```

Строим график функции

```python
plt.plot(x, np.arctan(x))
plt.grid(color="b", linewidth = "0.1")
plt.xlabel("x")
plt.ylabel("Arctg(x)")
plt.show()
```

Считаем первую производную функции и строим её график аналитически и способом приращений

```python
f1a = 1/(1+x**2)
f1 = (np.arctan(x)-np.arctan(x-h))/h
sp = plt.subplot(221)
plt.plot(x, f1a)
plt.grid(color="b", linewidth = "0.1")
plt.xlabel("x")
plt.ylabel("f2")

sp = plt.subplot(222)
plt.plot(x, f1, 'k-')
plt.grid(color="b", linewidth = "0.1")
plt.xlabel("x")
plt.ylabel("f2")
plt.show()
```

Наложение графиков

```python
plt.plot(x, f1a, x, f1)
plt.grid(color="b", linewidth = "0.1")
plt.xlabel("x")
plt.ylabel("f2")
plt.show()
```

Считаем вторую производную функции и строим её график аналитически и способом приращений

```python
f2a = (-2*x)/((1+x**2)**2)
f2 = (np.arctan(x)-2*np.arctan(x-h)+np.arctan(x-2*h))/(h**2)
sp = plt.subplot(221)
plt.plot(x, f2a)
plt.grid(color="b", linewidth = "0.1")
plt.xlabel("x")
plt.ylabel("f2")

sp = plt.subplot(222)
plt.plot(x, f2, 'k-')
plt.grid(color="b", linewidth = "0.1")
plt.xlabel("x")
plt.ylabel("f2")
plt.show()
```

Наложение графиков

```python
plt.plot(x, f2a, x, f2)
plt.grid(color="b", linewidth = "0.1")
plt.xlabel("x")
plt.ylabel("f2")
plt.show()
```

Строим график разности второй производной взятой с помощью приращений b второй производной взятой аналитически 

```python
plt.plot(x, f2-f2a, 'r-')
plt.grid(color="b", linewidth = "0.1")
plt.xlabel("x")
plt.ylabel("f2")
plt.show()
```
